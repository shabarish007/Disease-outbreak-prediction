```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
heart_data = pd.read_csv('/content/heart.csv')
```

```python
print(heart_data.columns)
```

```python
x=heart_data.drop(columns='target',axis=1)
y=heart_data['target']
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

**Support Vector Machine**

```python
svmmodel=SVC(kernel='linear')
svmmodel.fit(x_train,y_train)
```

```python
train_pred=svmmodel.predict(x_train)
train_acc=accuracy_score(train_pred,y_train)
print("Training Accuracy:",train_acc)
```

```python
test_pred=svmmodel.predict(x_test)
test_acc=accuracy_score(test_pred,y_test)
print("Test Accuracy:",test_acc)
```

**LOGISTIC REGRESSION**

```python
logistic_model=LogisticRegression(max_iter=500)
logistic_model.fit(x_train,y_train)
```

```python
train_pred=logistic_model.predict(x_train)
train_acc=accuracy_score(train_pred,y_train)
print("Training Accuracy:",train_acc)
```

```python
test_pred=logistic_model.predict(x_test)
test_acc=accuracy_score(test_pred,y_test)
print("Test Accuracy:",test_acc)
```

**DECISION TREE CLASSIFIER**

```python
tree_model=DecisionTreeClassifier(criterion='gini')
tree_model.fit(x_train,y_train)
```

```python
train_pred=tree_model.predict(x_train)
train_acc=accuracy_score(train_pred,y_train)
print("Training Accuracy:",train_acc)
```

```python
test_pred=tree_model.predict(x_test)
test_acc=accuracy_score(test_pred,y_test)
print("Test Accuracy:",test_acc)
```

**RANDOM FOREST CLASSIFIER**

```python
forest_model=RandomForestClassifier(n_estimators=100)
forest_model.fit(x_train,y_train)
```

```python
train_pred=forest_model.predict(x_train)
train_acc=accuracy_score(train_pred,y_train)
print("Training Accuracy:",train_acc)
```

```python
test_pred=forest_model.predict(x_test)
test_acc=accuracy_score(test_pred,y_test)
print("Test Accuracy:",test_acc)
```
